```python
import matplotlib.pyplot as plt 
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color="red")
```

```python
#y = mx +c1 this is linear regrassions basic  formula
```

```python
#y=m1x1+c1, m2x2+c2,m3X3 +c3  # for multi values
```

```python
#y=mx+c
#y=(prediction value or target or dependent variable ) for ex: price
#m= slop
#x=(independent variablr) for ex: area
#c=y-intercept
```

```python
price = df.price
```

```python
price
```

```python
x = df.drop('price' , axis = 'columns') # x ko humne price m leliya or y ko area m formula k hisab se 
```

```python
x
```

```python
df[['area']] # asa bhi le sakte hen or drop se bhi
```

```python
from sklearn import linear_model
```

```python
## linear model = linear regrassion / logistic regrassion etc
```

```python
# creating linear model object
reg = linear_model.LinearRegression()
```

```python
reg.fit(x,price)# prediction data 2 diamission m data dengy or fit m hum data fit kar rahe hen
```

```python
reg.predict([[5000]]) # 5000 k area k hisab se price predict kar k dega
```

```python
reg.predict([[10000]])
```
